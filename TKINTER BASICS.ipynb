```python
#Importing Button, Entry and label and pacing them
import tkinter as tk
from tkinter import *
window=tk.Tk()
window.geometry("500x500")
window.title("my EMI Tkinter")
lblp= Label(window, text="Enter principal")
lblp.grid(column=0, row=1)
p=Entry(window)
p.grid(column=0, row=2)
btnsi=Button(window, text="Calculate SI", bg='blue', fg='white')
btnsi.grid(column=0, row=3)
window.mainloop()
```

```python
#Importing Button, Entry and label and pacing them
import tkinter as tk
from tkinter import *
window=tk.Tk()
window.geometry("500x500")
window.title("my EMI Tkinter")

#label:p
lblp= Label(window, text="Enter principal")
lblp.grid(column=0, row=1)
p=Entry(window)
p.grid(column=1, row=1)

#label:r
lblr= Label(window, text="Enter rate")
lblr.grid(column=0, row=3)
r=Entry(window)
r.grid(column=1, row=3)

#label:t
lblt= Label(window, text="Enter time")
lblt.grid(column=0, row=5)
t=Entry(window)
t.grid(column=1, row=5)

#button SI
btnsi=Button(window, text="Calculate SI", bg='blue', fg='white')
btnsi.grid(column=1, row=7)
window.mainloop()
```

```python
#using pack/place 

import tkinter as tk
from tkinter import *
window=tk.Tk()
window.geometry("500x500")
window.title("my EMI Tkinter")


lblp= Label(window, text="Enter principal")
#lblp.pack()
lblp.place(x=200,y=200)
p=Entry(window)

p.pack()

btnsi=Button(window, text="Calculate SI", bg='blue', fg='white')
btnsi.pack()


window.mainloop()
```

```python
import tkinter as tk
from tkinter import *
window=tk.Tk()
window.geometry("500x500")

def cal():
    #print(p.get())  # string
    pr=float(p.get())*50
    lblresult.configure(text="calculated output is" + str(pr))

lblp= Label(window, text="Enter principal")
lblp.pack()

p=Entry(window)
p.pack()

btnsi=Button(window,command=cal, text="Calculate SI", bg='blue', fg='white')
btnsi.pack()

lblresult=Label(window, text="result is")
lblresult.pack()

window.mainloop()
```
